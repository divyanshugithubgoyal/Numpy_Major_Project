```python
import numpy as np
```

```python

def arithmetic_operations():
    print("=== ARRAY OPERATIONS ===")
    try:
        arr1 = list(map(float, input("Enter elements of first array (space-separated): ").split()))
        arr2 = list(map(float, input("Enter elements of second array (space-separated): ").split()))
        a = np.array(arr1)
        b = np.array(arr2)

        if len(a) != len(b):
            print("Error: Arrays must have the same length for element-wise operations.")
            return

        print("\n--- Results ---")
        print("Addition:", a + b)
        print("Subtraction:", a - b)
        print("Multiplication:", a * b)
        print("Division:", a / b)
        print("Power:", a ** b)

    except Exception as e:
        print("An error occurred:", e)

def statistical_operations():
    print("=== STATISTICAL OPERATIONS ===")
    try:
        arr = list(map(float, input("Enter elements of the array (space-separated): ").split()))
        a = np.array(arr)

        print("\n--- Results ---")
        print("Array:", a)
        print("Mean:", np.mean(a))
        print("Median:", np.median(a))
        print("Minimum:", np.min(a))
        print("Maximum:", np.max(a))
        print("Sum:", np.sum(a))
        print("Product:", np.prod(a))
        print("Standard Deviation:", np.std(a))
        print("Variance:", np.var(a))
        print("Shape:", a.shape)
        print("Size:", a.size)
        print("Data Type:", a.dtype)
    except Exception as e:
        print("An error occurred:", e)

def array_manipulation():
    print("=== ARRAY MANIPULATION WITH ALL POSSIBLE RESHAPES ===")
    try:
        arr = list(map(float, input("Enter elements of the array (space-separated): ").split()))
        a = np.array(arr)
        print("\nOriginal Array:", a)
        n = a.size

        print("\n All Possible 2D Reshapes ")
        for i in range(1, n + 1):
            if n % i == 0:
                try:
                    print(f"{i} x {n//i}:\n", a.reshape(i, n//i))
                except:
                    pass


        print("\n All Possible 3D Reshapes ")
        for i in range(1, n + 1):
            for j in range(1, n + 1):
                if n % (i * j) == 0:
                    try:
                        print(f"{i} x {j} x {n//(i*j)}:\n", a.reshape(i, j, n//(i*j)))
                    except:
                        pass


        flat_a = a.flatten()


        sorted_array = np.sort(flat_a)
        print("\nSorted Array:", sorted_array)
        print("Reverse Sorted Array:", sorted_array[::-1])
        print("Unique Values:", np.unique(flat_a))


        print("First Element:", flat_a[0])
        print("Last Element:", flat_a[-1])


        mean_val = np.mean(flat_a)
        greater_than_mean = flat_a[flat_a > mean_val]
        print("Values greater than mean ({}):".format(mean_val), greater_than_mean)

    except Exception as e:
        print("An error occurred:", e)

def exit_program():
    print("Exiting the program. Goodbye!")
    return True



while True:
    print("\n======================================================")
    print("Enter zero to execute Arithmetic Operations ---- 1")
    print("Enter One to execute Statistical Operations ---- 2")
    print("Enter Two to execute Array Manipulation ---- 3")
    print("Enter Three to Exit the loop ---- 4")
    print("======================================================")
    try:
        choice = int(input("Enter your choice: "))

        if choice == 1:
            arithmetic_operations()
        elif choice == 2:
            statistical_operations()
        elif choice == 3:
            array_manipulation()
        elif choice == 4:
            if exit_program():
                break
        else:
            print("Invalid choice! Please enter a number between 1 and 4.")

    except Exception as e:
        print("Error:", e)
```
